# Cleanup total unemployment rate
The table currently contains extra information we are not interested in. At the end we ouput a cleaned file.

```python
import os

import pandas as pd

BASE = os.path.join(os.pardir, "data")
```

```python
# Open the file
df = pd.read_pickle(os.path.join(BASE, 
                                 "pickles", 
                                 "tps00203.pkl"))
df.head()
```

```python
# Check the different values present
df["unit"].unique()
```

```python
# Drop the units we are not interested in
df.drop(df[df["unit"] != "PC_ACT"].index, inplace=True)
```

```python
df["sex"].unique()
```

```python
# Three columns now contain a single value, so we get rid of those
df = (df
      .drop(columns=["age", "unit", "sex"])
      .set_index("geo"))
df.head()
```

```python
# Transpose
df = df.T

# Change columns for rows
df.columns.name = ""
df.head()
```

```python
# Make index as type integeer
df.index = df.index.astype(int)
```

```python
# Export
df.to_pickle(os.path.join(BASE, "pickles", "total_unemployment_rate.pkl"))
df.to_csv(os.path.join(BASE, "pickles", "total_unemployment_rate.csv"))
```

```python
def simplify_col(col):
    """Return column as series with dtype float."""
    
    def _split_str(x):
        x = x.split()
        return x[0] if isinstance(x, list) else x
    
    col = (np.where(col.str.contains(":"), 
                        pd.NA, 
                        col
                        .apply(_split_str)))    
    return pd.to_numeric(col, errors="coerce")
```
